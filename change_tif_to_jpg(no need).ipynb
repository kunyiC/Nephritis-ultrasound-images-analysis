```python
import os 
from PIL import Image 
```

```python
base_dir = '/home/ai03/_kunyi/pro_tasi/yolov5_4_v2/datasets/'
```

```python
img_source_dir = base_dir + "images/dataset_no_UB"
target_dir = base_dir + "images/dataset_no_UB_2"
```

```python
import pathlib

def check_if_file_exist(target_dir):
    file = pathlib.Path(target_dir)
    if file.exists():
        print ("File exist")
    else:
        print ("File not exist")
        os.makedirs(target_dir)
```

```python
check_if_file_exist(target_dir)
```

```python
files_list = os.listdir(img_source_dir)
```

```python
files_list
```

```python
print(files_list[0])
print(files_list[0][0:-3])
```

```python
files_dir = [os.path.join(
    img_source_dir, name) for name in files_list]
```

```python
files_dir
```

```python
for img_file, img_name in zip(files_dir, files_list):
    if img_file.endswith('.tif'):
        image = Image.open(img_file)
        #image_copied = image.copy()
        image.save(
            target_dir+"/"+img_name[0:-3]+'jpg')
```
